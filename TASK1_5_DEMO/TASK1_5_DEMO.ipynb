### Задача 1. Доля выживших

### Самое популярное женское имя

Какое самое популярное женское имя на корабле? Извлеките из полного имени пассажира (колонка Name) его имя (First Name)

```python
# создаем список с женскими именами

fem_names = list(df[df["Sex"] == "female"].Name)
fem_names
```

```python
# парсим женские имена
# если имя содержит "Mrs." берем имя в скобках
# если Miss. берем имя после точки
# то есть, общее правило - берем имена после точки
# а если что-то указано в скобка, то после скобки

# обновляем список перед парсингом

fem_names = list(df[df["Sex"] == "female"].Name)

# перебираем значения в списке женских имен

for i in range(len(fem_names)):
  #print(fem_names[i])
  if "(" in fem_names[i]: # если в имени содержится откр. скобка,
    # то разделяем имя на две части по скобке, берем вторую часть
    # и убираем закрывающую скобку
    fem_names[i] = fem_names[i].split("(")[1].replace(")", "")
  elif "Miss." in fem_names[i]: # если в имени содержится Miss.,
    # то разделяем имя на две части по точке с пробелом
    # и берем вторую часть.
    fem_names[i] = fem_names[i].split("Miss. ")[1]
  elif "Mrs. " in fem_names[i]: # если в имени содержится Mrs.,
    # то разделяем имя на две части по точке с пробелом
    # и берем вторую часть.
    fem_names[i] = fem_names[i].split("Mrs. ")[1]
  elif "Mme. " in fem_names[i]: # если в имени содержится Mme.,
    # то разделяем имя на две части по точке с пробелом
    # и берем вторую часть.
    fem_names[i] = fem_names[i].split("Mme. ")[1]
  elif "Mlle. " in fem_names[i]: # если в имени содержится Mlle.,
    # то разделяем имя на две части по точке с пробелом
    # и берем вторую часть.
    fem_names[i] = fem_names[i].split("Mlle. ")[1]
  elif "Ms. " in fem_names[i]: # если в имени содержится Ms.,
    # то разделяем имя на две части по точке с пробелом
    # и берем вторую часть.
    fem_names[i] = fem_names[i].split("Ms. ")[1]     
  #else:
    # для проверки в неизмененные строки подставляем строку дефисов
  #  fem_names[i] = "------------" + fem_names[i]   

  # убираем имена в кавычках напр. из Helen Mary "Ellie"
  #if '"' in fem_names[i]: # если в имени содержится двойная кавычка
    # разделяем имя на две части по пробелу и кавычке
    # и берем первую часть
  #  fem_names[i] = "------------" + fem_names[i].split(' "')[0]

  # Убираем двойные кавычки
  fem_names[i] = fem_names[i].replace('"', "")

fem_names
```

```python
# собираем в словарь имена из обработанного списка fem_names

fem_name_dict = {} # объявляем словарь

# перебираем все имена из списка fem_names

for name in fem_names:
  for name_part in name.split(" "): # берем первое слово из разделенного имени
    # если данное слово есть в ключах нашего словаря
    if name_part in fem_name_dict:
      # увеличиваем счетчик имени в словаре на 1
      fem_name_dict[name_part] += 1
    else: # если имени ещё нет в ключах словаря fem_name_dict
      # добавляем в словарь новую пару ключ-значение
      fem_name_dict[name_part] = 1

fem_name_dict
```

```python
# выводим значения имен, у которых 9 и более повторений

for key, value in fem_name_dict.items():
  if value > 9: # если значение больше 9
    # то выводим на экран пару ключ-значение
    print(key, value)
```
